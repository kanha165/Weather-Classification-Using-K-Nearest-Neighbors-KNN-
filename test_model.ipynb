```python
import numpy as np
import pickle
```

```python
model = pickle.load(open("knn_weather_model.pkl", "rb"))
scaler = pickle.load(open("scaler.pkl", "rb"))
encoder = pickle.load(open("encoder.pkl", "rb"))
```

```python
temp = float(input("Enter Temperature: "))
humidity = float(input("Enter Humidity: "))
wind = float(input("Enter Wind Speed: "))
precip = float(input("Enter Precipitation (%): "))
pressure = float(input("Enter Atmospheric Pressure: "))
uv = float(input("Enter UV Index: "))
visibility = float(input("Enter Visibility (km): "))


user_data = np.array([[temp, humidity, wind, precip, pressure, uv, visibility]])


user_data = scaler.transform(user_data)

prediction = model.predict(user_data)
weather = encoder.inverse_transform(prediction)


print("\nðŸŒ¤ Predicted Weather Type:", weather[0])
```
